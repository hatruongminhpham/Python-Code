## Join Examples

```python
drinks = Table(['Drink', 'Cafe', 'Price']).with_rows([
    ['Milk Tea', 'Tea One', 4],
    ['Espresso', 'Nefeli',  2],
    ['Latte',    'Nefeli',  3],
    ['Espresso', "Abe's",   2]
])
drinks
```

```python
discounts = Table().with_columns(
    'Coupon % off', make_array(25, 50, 5),
    'Location', make_array('Tea One', 'Nefeli', 'Tea One')
)
discounts
```

```python
combined = drinks.join('Cafe', discounts, 'Location')
combined
```

```python
discounted_prices = combined.column('Price') * (1 - combined.column('Coupon % off')/ 100)
combined.with_column('Discounted', discounted_prices)
```

## Join Examples: The Fama-French 5-Factor Model

```python
the_four = Table.read_table('https://www.dropbox.com/s/fbfrt3ot9k3vrqc/the_four.csv?raw=1')
the_four
```

```python
ff = Table.read_table('https://www.dropbox.com/s/qtgk1urdf78tite/F-F_Research_Data_5_Factors_2x3.CSV?raw=1')
ff.show()
```

```python
combo = the_four.join('date', ff, 'Date')
combo
```

```python
# convert rtn to percents
# append_column() appends a column to the table or replaces a column
combo = combo.append_column('rtn', combo.column('rtn') * 100)
combo.show()
```

```python
amzn = combo.where('ticker', 'AMZN')
amzn
```

```python
# calculate excess returns
amzn = amzn.with_column('excess_rtn', amzn.column('rtn') - amzn.column('RF'))
amzn
```

```python
amzn.scatter('Mkt-RF', 'excess_rtn')
```

```python
amzn.scatter('SMB', 'excess_rtn')
```

```python
amzn.scatter('HML', 'excess_rtn')
```
